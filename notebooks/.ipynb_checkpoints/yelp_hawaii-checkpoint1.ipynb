```python
from bs4 import BeautifulSoup
import requests
```

```python
# URL of page to be scraped
url = "https://www.yelp.ca/search?find_desc=Restaurants&find_loc=Honolulu%2C+HI%2C+US&ns=1"
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all("a", class_="lemon--a__373c0__1_OnJ link__373c0__29943 link-color--blue-dark__373c0__1mhJo link-size--inherit__373c0__2JXk5")
#<a class="lemon--a__373c0__1_OnJ link__373c0__29943 link-color--blue-dark__373c0__1mhJo link-size--inherit__373c0__2JXk5" href="/biz/helenas-hawaiian-food-honolulu?osq=Restaurants" target="" name="" rel="">Helena’s Hawaiian Food</a>
```

```python
# Loop through returned results
title = []
for result in results:
    # Error handling
    try:
        # Identify and return name of island
        name = result.text
        
        # Print results only if title, price, and link are available
        if (name):
            print('-------------')
            print(name)
            title.append(name)
    except AttributeError as e:
        print(e)
```
